# Stack
<br>

```python
import pandas as pd
import numpy as np
```

```python
 df = pd.DataFrame({
     'Province' : ['ON','QC','BC','AL','AL','MN','ON'],
     'City' : ['Toronto','Montreal','Vancouver','Calgary','Edmonton','Winnipeg','Windsor'],
     'Sales' : [13,6,16,8,4,3,1]
 })
```

```python
df
```

```python
table = pd.pivot_table(df,index=['Province'],columns=['City'], values=['Sales'],aggfunc=np.sum,margins=True)
```

```python
table
```

```python
stack_df = table.stack('City')
```

```python
stack_df.reset_index()
```

```python
stack_df
```

# real example

```python
df = pd.read_table('m_index.txt')
```

```python
df
```

```python
grp_ser = df.groupby(['LOT','WF','ITEM','STEP'])['VALUE'].agg(lambda x: x.mean())
```

```python
print(type(grp_ser))
print(grp_ser)
```

```python
grp_ser.reset_index()
```

```python
grp_df = df.groupby(['LOT','WF','ITEM','STEP'])[['VALUE']].agg(lambda x: x.mean())
```

```python
print(type(grp_df))
print(grp_df)
```

```python
grp_df.reset_index()
```
