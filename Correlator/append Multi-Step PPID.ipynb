```python
res = pd.merge(measure, crit, on=['step'], how='left')

res
```

```python
res2 = res2.reset_index()
```

```python
res2
```

```python
res2.rename(columns={0:'step_p','level_4':'step_info'}, inplace=True)
```

```python
res2
```

```python
res3 = pd.merge(res2, path, on=['lot_wf','step_p'], how='left')

res3
```

```python
pivot = pd.pivot_table(
    res3,
    index=['step','lot_wf'],
    columns=['step_info'],
    values='ppid_p',
    aggfunc='last',
).reset_index()

pivot
```

### append conditional information

```python
pivot.loc[pivot['step_p2'].isnull(), 'alias'] = pivot['step_p']
pivot.loc[pivot['step_p2'].notnull(), 'alias'] = pivot['step_p'] + '_' + pivot['step_p2']

pivot
```
